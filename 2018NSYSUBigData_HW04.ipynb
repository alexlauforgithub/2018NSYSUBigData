# OOP normalize scaler設計

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
class MyScaler:
        
    def __init__(self):
        None
        
    def fit(self,x):
        self.mu = np.mean(x)
        self.std = np.std(x)
    
    def transform(self,x):
        return((x-self.mu)/self.std)
        
```

## 隨機產生mu=10, std=10的常態隨機變數

```python
mu, sigma = 10, 10
y = np.random.normal(mu, sigma, size = (1000))
print(y[1:20])
```

```python
fig, ax=plt.subplots()
sns.distplot(y, ax=ax)
```

## 建立instance

```python
scaler = MyScaler()
```

## 將原數據進行標準化並畫圖

```python
scaler.fit(y)
```

```python
fig, ax=plt.subplots()
sns.distplot(scaler.transform(y), ax=ax)
```

## 產生新的mu=10, std=10的常態隨機變數，並套入原有scaler進行標準化並畫圖

```python
mu, sigma = 10, 10
y_new = np.random.normal(mu, sigma, size = (1000))
```

```python
fig, ax=plt.subplots()
sns.distplot(scaler.transform(y_new), ax=ax)
```
